```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
file_path = Path('./Tables/job_final_Reason.csv')
job = pd.read_csv(file_path)
job.head()
```

```python
file_path = Path('./Tables/health_final.csv')
mental = pd.read_csv(file_path)
mental.head()
```

```python
final = pd.merge(mental, job, on=["REF_DATE", "GEO",'Sex','Age group'])
final.head(20)
```

```python
final.info()
```

```python
final.rename(columns={'REF_DATE': 'Year',
                        'GEO' : 'Province',
                        'Sex' : 'Sex',
                        'Age group' : 'Age_group',
                        'Indicators' : 'Indicators_Health',
                        'VALUE' : 'Num_Person_Health',
                        'Dissatisfied' : 'J_L_R_Dissatisfied',
                        'Going to school' : 'J_L_R_Going_to_school',
                        'Have not worked in last year' : 'J_L_R_not_worked_last_year',
                        'Job leavers' : 'J_L_R_Job_leavers',
                        'Jobs losers' : 'J_L_R_Jobs_losers',
                        'Never worked' : 'J_L_R_Never_worked',
                        'Other reasons' : 'J_L_R_Other_reasons',
                        'Own illness or disability' : 'J_L_R_Own_illness_disability',
                        'Permanent layoff' : 'J_L_R_Permanent_layoff',
                        'Personal or family reasons' : 'J_L_R_Personal_family_reasons',
                        'Retired' : 'J_L_R_Retired',
                        'Temporary layoff' : 'J_L_R_Temporary_layoff',
}, inplace=True)
```

```python
final.head(8)
```

```python
final.to_csv('./Tables/healthXjob_final.csv', index=True)
```
