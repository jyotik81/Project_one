```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
file_path = Path('./Tables/health/13100096.csv')
mental = pd.read_csv(file_path)
print(mental.shape)
mental.head()
```

```python
# clean the unnecessary collumns
mental = mental.drop(['DGUID','UOM','UOM_ID','SCALAR_FACTOR','SCALAR_ID','VECTOR','COORDINATE','STATUS','SYMBOL','TERMINATED','DECIMALS'], axis=1)
print(mental.shape)
mental.head()
```

```python
#find null values
for column in mental.columns:
    print(f'{column} has {mental[column].isnull().sum()} null values')
```

```python
# Remove rows that have at least 1 null value.
mental = mental.dropna()
print(mental.shape)
mental.head(10)
```

```python
#Filter by Characteristics
caract_filter = mental['Characteristics'] == 'Number of persons'
mental = mental.loc[caract_filter]
mental = mental.drop(['Characteristics'], axis=1)
```

```python
#Filter by Sex - Exclude both sexes
sex_filter = mental['Sex'] != 'Both sexes'
mental = mental.loc[sex_filter]
```

```python
#Filter only Provinces
geo_filter = mental['GEO'] != 'Canada (excluding territories)'
mental = mental.loc[geo_filter]
```

```python
#Filter date between 2002 and 2012
date_filter = (mental['REF_DATE']>2016)
mental = mental.loc[date_filter]
mental.head()
```

```python
mental['Age group'].value_counts()
```

```python
#Filter Age
age_filter = mental['Age group'].isin(['18 to 34 years', '35 to 49 years','50 to 64 years','65 years and over'])
mental = mental.loc[age_filter]
mental.head(20)
```

```python
mental['Age group']=mental['Age group'].replace({'18 to 34 years': 1})
mental['Age group']=mental['Age group'].replace({'35 to 49 years': 2})
mental['Age group']=mental['Age group'].replace({'50 to 64 years': 3})
mental['Age group']=mental['Age group'].replace({'65 years and over': 4})
mental.head(10)
```

```python
mental['VALUE'] = mental['VALUE'].fillna(0).astype(int)
```

```python
#Filter Reason 'Permanent layoff'
mental_final = mental.groupby(['REF_DATE','GEO','Sex','Age group','Indicators'])['VALUE'].sum()
mental_final = pd.DataFrame(mental_final)
mental_final.head(20)
```

```python
#Pivot Reason variable
mental_final = mental_final.pivot_table(index=['REF_DATE','GEO','Sex','Age group'],columns='Indicators',values='VALUE')
mental_final.head(10)
```
